```python
import pandas as pd
import numpy as np
from datetime import datetime
```

# Reading in the Kaggle Data

```python
df = pd.read_csv('../datasets/kaggle_harvey_tweets_2.csv')
df.head()
```

```python
df.drop(columns = 'Unnamed: 0', inplace = True)
df.head()
```

```python
df_2 = pd.read_csv('../datasets/kaggle_hurricane_harvey.csv')
df_2.head()
```

```python
df_2.drop(columns = 'Unnamed: 0', inplace = True)
df_2.head()
```

```python
df.shape
```

```python
df_2.shape
```

```python
df_kaggle = pd.concat([df, df_2], axis=0)
df_kaggle.head()
```

```python
df_kaggle.shape
```

# Dropping Null Values

```python
df_kaggle = df_kaggle.dropna()
df_kaggle = df_kaggle.reset_index(drop=True)
df_kaggle.head()
```

# Converting to Datetime Format

```python
df_kaggle['Time'] = pd.to_datetime(df_kaggle['Time'])

df_kaggle['date'] = df_kaggle['Time'].dt.date
df_kaggle.head(1)
```

```python
df_kaggle.drop(columns = ['Likes', 'Replies', 'Retweets', 'Time'], inplace = True)
df_kaggle.head(1)
```

```python
df_kaggle.columns = df_kaggle.columns.str.lower()
df_kaggle.head(1)
```

```python
df_kaggle.shape
```

# Dropping Duplicates

```python
df_kaggle.drop_duplicates(subset ="tweet", 
                     keep = False, inplace = True) 
df_kaggle.shape
```

# Reading in NTU Data

```python
df_3 = pd.read_csv('../datasets/ntu_twitter_1.csv')
df_3.head()
```

```python
df_4 = pd.read_csv('../datasets/ntu_twitter_2.csv')
df_4.head()
```

```python
df_5 = pd.read_csv('../datasets/ntu_twitter_3.csv')
df_5.head()
```

```python
df_6 = pd.read_csv('../datasets/ntu_twitter_4.csv')
df_6.head()
```

```python
df_ntu = pd.concat([df_3, df_4, df_5, df_6], axis = 0)
df_ntu.head()
```

```python
df_ntu.shape
```

# Dropping NA's

```python
df_ntu = df_ntu.dropna()
df_ntu = df_ntu.reset_index(drop=True)
df_ntu.head()
```

# Cleaning Up Column Names

```python
df_ntu.drop(columns = ['handle', 'name', 'replies', 'retweets', 'favorites', 'unix_timestamp', '__url'],
            inplace = True)
df_ntu.head(1)
```

```python
df_ntu.rename(columns = {'content': 'tweet', 'url': 'id'}, inplace = True)
df_ntu.head(1)
```

```python
df_ntu = df_ntu[['id', 'date', 'tweet']]
df_ntu.head()
```

```python
df_ntu.shape
```

# Dropping Duplicates

```python
df_ntu.drop_duplicates(subset ="tweet", 
                     keep = False, inplace = True) 
df_ntu.shape
```

```python
df_ntu.isnull().sum()
```

```python
df_kaggle.isnull().sum()
```

# Combining all Dataframes into Single Dataframe

```python
df_combined = pd.concat([df_kaggle, df_ntu])
df_combined.head()
```

```python
df_combined.shape
```

```python
df_combined.isnull().sum()
```

# Saving combined dataframe

```python
df_combined.to_csv('../datasets/df_combined.csv', index = False)
```
